```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing = pd.read_csv('housing.csv' , encoding = 'UTF-8')
```

```python
housing
```

```python
housing.columns
```

```python
housing['Rooms'].unique()
```

```python
housing.info()
```

```python
housing.dtypes
```

```python
sns.set_theme()

sns.relplot(data = housing , x = 'Sale Condition' , y = 'SalePrice')

plt.show()
```

```python
sns.set_theme()

sns.relplot(data = housing , x = 'Sale Condition' , y = 'SalePrice' , hue = 'Year Built', palette = 'RdYlBu')

plt.show()
```

```python
sns.set_theme()

sns.relplot(data = housing , x = 'Sale Condition' , y = 'SalePrice' , hue = 'Year Built', palette = 'RdYlBu',
           size ='Gr Liv Area')

plt.show()
```

```python
sns.set_theme()

sns.relplot(data = housing , x = 'Sale Condition' , y = 'SalePrice' , hue = 'Year Built', palette = 'RdYlBu',
           size ='Gr Liv Area' , sizes= (1,500) , style = 'Street')

plt.show()
```

```python
sns.set_theme()

sns.relplot(data = housing , x = 'Sale Condition' , y = 'SalePrice' , hue = 'Year Built', palette = 'RdYlBu',
           size ='Gr Liv Area' , sizes= (1,500) , style = 'Street' , col = 'Foundation' , col_wrap = 2)

plt.show()
```
